```python
TAX_RATE = 0.1 # 消費税率10%

try:
    price = int(input("税抜価格を入力してください: "))
except ValueError:
    print("数値を入力してください。")
except Exception as e:
    # その他エラー
    print(f"予期しないエラーが発生しました: {e}")
else:
    price_with_tax = price * (1 + TAX_RATE)
    
    shipping_cost = 0
    if price < 2000:
        shipping_cost = 350

    total = price_with_tax + shipping_cost
    print(f"送料は{shipping_cost if shipping_cost != 0 else '無料'}です。")
    print(f"送料込税込価格は{int(total)}です")
```
