```python
AMOUNT_YEARS = 10

try:
    principal = int(input("元金を入力してください: "))
    annual_rate = int(input("年利を入力してください: "))
except ValueError:
    print("数値を入力してください。")
except Exception as e:
    # その他エラー
    print(f"予期しないエラーが発生しました: {e}")
else:
    amount = principal

    i = 0
    while i < AMOUNT_YEARS:
        amount = amount * (1 + annual_rate / 100)
        i += 1
    print(f"10年後の金額:{amount:.2f}です。")
```
