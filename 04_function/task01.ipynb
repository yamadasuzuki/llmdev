```python
def is_valid(s):
    # ユーザー名は5文字以上15文字以内。条件を満たさない場合
    if len(s) < 5 or len(s) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    # ユーザー名に空白が含まれている場合
    if " " in s:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    # ユーザー名は英数字以外が含まれている場合
    if not s.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    return True

some_string = input("")
print("ユーザー名は有効です") if is_valid(some_string) else print("不正なユーザー名です")
```
