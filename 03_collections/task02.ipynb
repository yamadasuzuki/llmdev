```python
# 空文字が来るまでinputを繰り返す
strings = []
while True:
    tempStr = input("文字列を入力してください（終了するには空文字を入力）: ")
    if tempStr == "":
        break
    strings.append(tempStr)

print(f"入力した英単語： {strings}")

# a-zの文字が何回出現するかを数える
counts = {
    "a": 0,
    "b": 0,
    "c": 0,
    "d": 0,
    "e": 0,
    "f": 0,
    "g": 0,
    "h": 0,
    "i": 0,
    "j": 0,
    "k": 0,
    "l": 0,
    "m": 0,
    "n": 0,
    "o": 0,
    "p": 0,
    "q": 0,
    "r": 0,
    "s": 0,
    "t": 0,
    "u": 0,
    "v": 0,
    "w": 0,
    "x": 0,
    "y": 0,
    "z": 0,
}

for string in strings:
    for char in string:
        if char in counts:
            counts[char] += 1

for char in counts.keys():
    print(f"{char}が{counts[char]}個ありました")
```
